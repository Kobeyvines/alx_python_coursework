Write Python code to create a violin plot visualising the distribution of 'Annual_yield' across different 'Elevation' ranges. Use the provided elevation range categories (Low: < 300m, Medium: 300m - 600m, High: > 600m) to categorise the data before plotting.  Examine the violin plot displaying the distribution of 'Annual Yield' across three elevation categories (Low, Medium, High). What insight does the violin plot provide regarding the relationship between elevation ranges and annual yield distribution?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Categorize Elevation into Low, Medium, and High
def categorize_elevation(elev):
    if elev < 300:
        return 'Low'
    elif elev <= 600:
        return 'Medium'
    else:
        return 'High'

agri_df['Elevation_Category'] = agri_df['Elevation'].apply(categorize_elevation)

# Step 2: Create the violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=agri_df, x='Elevation_Category', y='Annual_yield', palette='Set2')

plt.title('Distribution of Annual Yield Across Elevation Categories')
plt.xlabel('Elevation Range')
plt.ylabel('Annual Yield')
plt.show()
```

Assuming each 'Crop_type' contributes an integer value equal to its length (e.g., 'wheat' contributes 5), write a recursive function to sum the integer values for each unique crop type in the dataset. What is the sum?

```python
def crop_length_sum(crops):
    # Base case: if list is empty, return 0
    if not crops:
        return 0
    # Recursive case: length of first crop + sum of the rest
    return len(crops[0]) + crop_length_sum(crops[1:])

# Get unique crop types
unique_crops = agri_df['Crop_type'].unique().tolist()

# Calculate total sum
total_sum = crop_length_sum(unique_crops)
print(total_sum)
```

Write Python code to perform a t-test comparing the average 'Annual_yield' between 'coffee' and 'banana' crop types using scipy.stats. What is the p-value, rounded to three decimal places?

```python
from scipy import stats

# Step 1: Filter the two crop types
coffee = agri_df[agri_df['Crop_type'] == 'coffee']['Annual_yield']
banana = agri_df[agri_df['Crop_type'] == 'banana']['Annual_yield']

# Step 2: Perform an independent t-test
t_stat, p_value = stats.ttest_ind(coffee, banana, equal_var=False)  # Welch's t-test (safer if variances differ)

# Step 3: Round p-value to three decimal places
p_value_rounded = round(p_value, 3)

print("P-value:", p_value_rounded)
```
